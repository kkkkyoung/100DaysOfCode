```python
import random
#from reprlib import clear
from blackjack_logo import logo

def deal_card():
    """임의의 카드를 덱에서 반환"""
    cards = [11,2,3,4,5,6,7,8,9,10,10,10,10,]
    card = random.choice(cards)
    return card

def calculate_score(cards):
    """리스트를 가져와서 카드 점수를 반환"""
    if sum(cards) == 21 and len(cards) ==2 :
        return 0    # 블랙잭 : 10+11 카드 들고있을때
    if 11 in cards and sum(cards) > 21 : # 에이스 카드를 들고있으면 21점이 넘을땐 11점이 아닌 1점으로 계산
        cards.remove(11)
        cards.append(1)
    return sum(cards)

def compare(user_score, computer_score):
    """결과 판정"""
    if user_score == computer_score:
        return "무승부~~~"
    elif computer_score == 0:
        return "이런... 상대가 블랙잭이 나왔으므로 졌습니다ㅠㅠ"
    elif user_score == 0:
        return "아자~ 블랙잭! 승리!!!"
    elif user_score > 21:
        return "21점이 넘어버렸어요...졌습니다ㅠㅠ"
    elif computer_score > 21:
        return "상대가 21넘네요~ 나의 승리!!!"
    elif user_score > computer_score:
        return "아자! 나의 승리!!!"
    elif user_score > 21 and computer_score > 21:
        return "둘 다 21을 넘었으므로 당신이 졌습니다ㅠㅠ"
    else:
        return "졌습니다ㅠㅠ"

def play_game():
    """게임시작"""
    print(logo)
    user_cards = []
    computer_cards = []
    game_over = False

    for _ in range(2):
        user_cards.append(deal_card())  # 여러개를 넣을땐 += 사용가능 expend 함수 기능
        computer_cards.append(deal_card())

    while not game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)
        print(f"나의 카드는 : {user_cards}, 점수 : {user_score}")
        print(f"상대의 카드 한장 : {computer_cards[0]}")

        if user_score == 0 or computer_score ==0 or user_score > 21:
            game_over = True
        else:
            user_should_deal = input("다른 카드를 원한다면 'y', 아니라면 'n'를 입력하시오")
            if user_should_deal == "y":
                user_cards.append(deal_card())
            else:
                game_over = True

    while computer_score != 0 and computer_score < 17: # 블랙잭이 아니고 17이하일 때 컴퓨터가 카드 뽑기
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)

    print(compare(user_score, computer_score))
```

```python
while input("게임을 시작할까요? 'y' or 'n'") == "y":
#    clear()
    play_game()
```
