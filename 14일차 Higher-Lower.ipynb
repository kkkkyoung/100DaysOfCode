```python
from logo_vs_14 import logo, vs
from game_data_14 import data
import random
```

```python
# 인스타 팔로워수 비교 게임

def format_data(account):
    """계정을 넣으면-이름, 설명, 출신국가 반환"""
    account_name = account["name"]
    account_descr = account["description"]
    account_country = account["country"]
    return f"계정 이름:{account_name}, 설명:{account_descr}, 출신:{account_country}"

def check_answer(guess, a_follwer, b_follwer):
    """사용자의 추측에 대한 follwer수 결과값을 알려줌"""
    if a_follwer > b_follwer:
        return guess == "a"  # a의 팔로워가 크고 추측한 값도 a이면 true 반환
    else:
        return guess == "b"

print(logo)
score = 0
game_continue = True
account_b = random.choice(data)

while game_continue : 
    account_a = account_b  # 이전게임의 계정b 를 계정a 로 만들어서 게임 계속
    account_b = random.choice(data)
    
    while account_a == account_b:
        account_b == random.choice(data)

    print(f"비교대상 a : {format_data(account_a)}")
    print(vs)
    print(f"비교대상 b : {format_data(account_b)}")
    guess = input("누구의 팔로워 수가 더 많을것 같은가? 'a' or 'b' : ").lower()

    a_follwer_count = account_a["follower_count"]
    b_follwer_count = account_b["follower_count"]
    is_correct = check_answer(guess, a_follwer_count, b_follwer_count)

    if is_correct:
        score += 1
        print(f"정답입니다~! (당신의 현재 점수는 : {score})")
        print('*'*100)
    else:
        game_continue = False
        print(f"오답입니다ㅠ (당신의 현재 점수는 : {score})")
```
