```python
MENU = {
    "에스프레소": {
        "ingredients": {
            "water": 50,
            "coffee": 18,
        },
        "cost": 1500,
    },
    "라떼": {
        "ingredients": {
            "water": 200,
            "milk": 150,
            "coffee": 24,
        },
        "cost": 2000,
    },
    "카푸치노": {
        "ingredients": {
            "water": 250,
            "milk": 100,
            "coffee": 24,
        },
        "cost": 2500,
    }
}

profit = 0
resources = {
    "water": 300,
    "milk": 200,
    "coffee": 100,
}
```

```python
def is_resource_sufficient(order_ingredients):
    """주문이 가능하면 true, 재료가 부족하면 false"""
    for item in order_ingredients:
        if order_ingredients[item] >= resources[item]:
            print(f"죄송합니다. {item}가 부족합니다")
            return False
    return True

def process_coins():
    """투입된 금액을 알려줍니다"""
    print("동전을 넣어주세요")
    total = int(input("50원 동전을 몇개 넣었나요?")) * 50
    total += int(input("100원 동전을 몇개 넣었나요?")) * 100
    total += int(input("500원 동전을 몇개 넣었나요?")) * 500
    return total

def is_transaction_successful(money_received, drink_cost):
    """금액이 충분하면 true, 부족하면 false"""
    if money_received >= drink_cost:
        change = round(money_received - drink_cost, 2)
        print(f"거스름돈은 {change}원 입니다")
        global profit
        profit += drink_cost
        return True
    else:
        print("투입된 돈이 충분하지 않습니다. 동전을 돌려드리겠습니다.")
        return False

def make_coffee(drink_name, order_ingredients):
    """사용된 재료양 만큼 자판기 재료 차감"""
    for item in order_ingredients:
        resources[item] -= order_ingredients[item]
    print(f"{drink_name} 나왔습니다~")
    
is_on = True

while is_on:
    choice = input("어떤걸 드시겠습니까? (에스프레소 / 라떼 / 카푸치노) : ")
    if choice == "off":
        is_on = False
    elif choice == "report":
        print(f"water : {resources['water']}ml")
        print(f"milk : {resources['milk']}ml")
        print(f"coffee : {resources['coffee']}g")
        pint(f"money : {profit}")
    elif choice not in list(MENU.keys()):
        print("다시 입력해주세요")
    else:
        drink = MENU[choice]
        if is_resource_sufficient(drink["ingredients"]):
            payment = process_coins()
            if is_transaction_successful(payment, drink["cost"]):
                make_coffee(choice, drink["ingredients"])
```
