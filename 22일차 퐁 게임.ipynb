```python
# 퐁 게임 만들기 1 - 패들만들고 움직여보기
from turtle import Screen, Turtle 

screen = Screen()
screen.bgcolor("black")
screen.setup(width=800, height=600)
screen.title("Pong Game")
screen.tracer(0)

paddle = Turtle()
paddle.shape("square")
paddle.color("white")
paddle.shapesize(stretch_wid=5, stretch_len=1)
paddle.penup()
paddle.goto(350, 0)

def go_up():
    new_y = paddle.ycor() + 20
    paddle.goto(paddle.xcor(), new_y)
    
def go_down():
    new_y = paddle.ycor() - 20
    paddle.goto(paddle.xcor(), new_y)
    
screen.listen()
screen.onkey(go_up, "Up")
screen.onkey(go_down, "Down")

geme_is_on =True
while geme_is_on:
    screen.update()

screen.exitonclick()
```

```python
# 퐁 게임 만들기 5 - 패들과의 충돌 감지

from turtle import Screen, Turtle 
from paddle import Paddle
from ball import Ball
import time

screen = Screen()
screen.bgcolor("black")
screen.setup(width=800, height=600)
screen.title("Pong Game")
screen.tracer(0)

r_paddle = Paddle((350, 0))
l_paddle = Paddle((-350, 0))
ball = Ball()
   
screen.listen()
screen.onkey(r_paddle.go_up, "Up")
screen.onkey(r_paddle.go_down, "Down")
screen.onkey(l_paddle.go_up, "w")
screen.onkey(l_paddle.go_down, "s")

geme_is_on =True
while geme_is_on:
    time.sleep(0.1)
    screen.update()
    ball.move()
    
    # 상하단 벽 충돌 감지
    if ball.ycor() > 280 or ball.ycor() < -280:
        ball.bounce_y()
        
    # 패들과의 충돌 감지
    if ball.distance(r_paddle) < 50 and ball.xcor() > 320 or ball.distance(l_paddle) < 50 and ball.xcor() < -320:
        ball.bounce_x()

screen.exitonclick()
```

```python
# 퐁 게임 만들기 6 - 패들이 공을 놓쳤는지

from turtle import Screen, Turtle 
from paddle import Paddle
from ball import Ball
import time

screen = Screen()
screen.bgcolor("black")
screen.setup(width=800, height=600)
screen.title("Pong Game")
screen.tracer(0)

r_paddle = Paddle((350, 0))
l_paddle = Paddle((-350, 0))
ball = Ball()
   
screen.listen()
screen.onkey(r_paddle.go_up, "Up")
screen.onkey(r_paddle.go_down, "Down")
screen.onkey(l_paddle.go_up, "w")
screen.onkey(l_paddle.go_down, "s")

geme_is_on =True
while geme_is_on:
    time.sleep(0.1)
    screen.update()
    ball.move()
    
    # 상하단 벽 충돌 감지
    if ball.ycor() > 280 or ball.ycor() < -280:
        ball.bounce_y()
        
    # 패들과의 충돌 감지
    if ball.distance(r_paddle) < 50 and ball.xcor() > 320 or ball.distance(l_paddle) < 50 and ball.xcor() < -320:
        ball.bounce_x()
        
    # 공을 놓쳤을떄
    if ball.xcor() > 380:
        ball.reset_position()
        
    if ball.xcor() < -380:
        ball.reset_position()
        
        
screen.exitonclick()
```

```python
# 퐁 게임 만들기 7 - 점수판 만들기, 공 속도 올리기

from turtle import Screen, Turtle 
from paddle import Paddle
from ball import Ball
import time
from scoreboard import Scoreboard

screen = Screen()
screen.bgcolor("black")
screen.setup(width=800, height=600)
screen.title("Pong Game")
screen.tracer(0)

r_paddle = Paddle((350, 0))
l_paddle = Paddle((-350, 0))
ball = Ball()
scoreboard = Scoreboard()
   
screen.listen()
screen.onkey(r_paddle.go_up, "Up")
screen.onkey(r_paddle.go_down, "Down")
screen.onkey(l_paddle.go_up, "w")
screen.onkey(l_paddle.go_down, "s")

geme_is_on =True
while geme_is_on:
    time.sleep(ball.move_speed)
    screen.update()
    ball.move()
    
    # 상하단 벽 충돌 감지
    if ball.ycor() > 280 or ball.ycor() < -280:
        ball.bounce_y()
        
    # 패들과의 충돌 감지
    if ball.distance(r_paddle) < 50 and ball.xcor() > 320 or ball.distance(l_paddle) < 50 and ball.xcor() < -320:
        ball.bounce_x()
        
    # 공을 놓쳤을떄
    if ball.xcor() > 380:
        ball.reset_position()
        scoreboard.l_point()
        
    if ball.xcor() < -380:
        ball.reset_position()
        scoreboard.r_point()
        
screen.exitonclick()
```
