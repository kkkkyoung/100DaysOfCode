```python
# 1 - 주가 변동 확인하기

import requests

STOCK_NAME = "TSLA"
COMPANY_NAME = "Tesla Inc"

STOCK_ENDPOINT = "https://www.alphavantage.co/query"
NEWS_ENDPOINT = "https://newsapi.org/v2/everything"

STOCK_API_KEY = 'my_STOCK_API_KEY'
stock_params={
    "function":"TIME_SERIES_DAILY_ADJUSTED",
    "symbol":STOCK_NAME,
    "interval":"30min",
    "apikey":STOCK_API_KEY,
}

response = requests.get(STOCK_ENDPOINT, params=stock_params)
# print(response.json())
data = response.json()['Time Series (Daily)']
data_list = [value for (key, value) in data.items()]
yesterday_data = data_list[0]
yesterday_closing_price = yesterday_data['4. close']
print(yesterday_closing_price) # 어제의 폐장가

day_befor_yesterday_data = data_list[1]
day_befor_yesterday_closing_price = day_befor_yesterday_data['4. close']
print(day_befor_yesterday_closing_price) # 그제의 폐장가

difference = abs(float(yesterday_closing_price) - float(day_befor_yesterday_closing_price))
print(difference) # 양의 차이값 (abs:절대값)

diff_percent = (difference / float(yesterday_closing_price)) * 100
print(diff_percent)

if diff_percent > 5:
    print("Get News")
```

```python
# 3 - sms메세지 보내기

import requests
from twilio.rest import Client

STOCK_NAME = "TSLA"
COMPANY_NAME = "Tesla Inc"

STOCK_ENDPOINT = "https://www.alphavantage.co/query"
NEWS_ENDPOINT = "https://newsapi.org/v2/everything"

STOCK_API_KEY = 'my_STOCK_API_KEY'
NEWS_API_KEY = 'my_NEWS_API_KEY'

account_sid = 'my_sid'
auth_token = 'my_token' 

stock_params={
    "function":"TIME_SERIES_DAILY_ADJUSTED",
    "symbol":STOCK_NAME,
    "interval":"30min",
    "apikey":STOCK_API_KEY,
}

response = requests.get(STOCK_ENDPOINT, params=stock_params)

data = response.json()['Time Series (Daily)']
data_list = [value for (key, value) in data.items()]
yesterday_data = data_list[0]
yesterday_closing_price = yesterday_data['4. close']
print(yesterday_closing_price) # 어제의 폐장가

day_befor_yesterday_data = data_list[1]
day_befor_yesterday_closing_price = day_befor_yesterday_data['4. close']
print(day_befor_yesterday_closing_price) # 그제의 폐장가

difference = abs(float(yesterday_closing_price) - float(day_befor_yesterday_closing_price))
print(difference) # 양의 차이값 (abs:절대값)

diff_percent = (difference / float(yesterday_closing_price)) * 100
print(diff_percent)

if diff_percent > 1:
    news_params= {
        "apiKey":NEWS_API_KEY,
        "q":COMPANY_NAME,
    }
    news_response = requests.get(NEWS_ENDPOINT, params=news_params)
    articles = news_response.json()["articles"]
    three_articles = articles[:3] # 기사 3개만 가져오기
    
    formatted_articles=[f"Healine:{article['title']}.\nBrief:{article['description']}" for article in three_articles]
    client = Client(account_sid, auth_token)
    for article in formatted_articles:
        message = client.messages.create(body=article,
                                         from_='보내는사람',
                                         to='받는사람')
```
